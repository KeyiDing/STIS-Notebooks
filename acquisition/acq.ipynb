```python
def acq(obj_id):
    raw = "/grp/hst/stis/calibration/monitors/monthly/acqs/data/data_recent/{}_raw.fits".format(obj_id)
    
#     plot the acquisition raw images
    with fits.open(raw) as hdu:
        initial = hdu[1].data
        confirm = hdu[4].data
        lamp = hdu[7].data
    plt.subplot(1,3,1)
    plt.imshow(initial,vmin = 0,vmax = 50)
    plt.title("Initial ACQ Image")
    plt.subplot(1,3,2)
    plt.imshow(confirm,vmin = 0,vmax = 50)
    plt.title("Confirmation ACQ Image")
    plt.subplot(1,3,3)
    plt.imshow(lamp,vmin = 0,vmax = 50)
    plt.title("HITM Lamp")
    plt.tight_layout()
    
    stistools.tastis.tastis(raw)
```
