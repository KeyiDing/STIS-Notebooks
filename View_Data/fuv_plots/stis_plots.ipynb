```python
from astropy.io import fits
import numpy as np
```

```python
stis_x1d = "o8k401010_x1d.fits"
hdul = fits.open(stis_x1d)
```

```python
hdul.info()
```

```python
fits.getheader(stis_x1d,1)
```

```python
stis_x1d_data = fits.getdata(stis_x1d,1)
```

```python
wl=stis_x1d_data["wavelength"].flatten()
wl
```

```python
flux=stis_x1d_data["flux"].flatten()
flux
```

```python
# Run this cell.
%matplotlib inline
import matplotlib
from matplotlib import pyplot as plt
matplotlib.rcParams['figure.figsize'] = (20,7)
plt.style.use("bmh")
```

```python
plt.plot(wl, flux, color="blue", label="STIS")
plt.legend(loc="best")
plt.xlabel("Wavelength [$\AA$]")
plt.ylabel("Flux [ergs/s/cm$^2$/$\AA$]")
```
